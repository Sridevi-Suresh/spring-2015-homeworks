To practice our filtering skills, let's try to isolate the cases that were opened in February **(5 pts)**

````python
# Filters the dataframe and keeps only the cases that were opened in February
rows = []
count = 0
for i in bos['OPEN_DT'].values[:]: 
    if not i == np.datetime64('NaT'):
        j = np.datetime_as_string(i, timezone='local')[:10]
        if j[5:7] == '02':
            rows.append(count)
    count += 1
drop_rows = []
for k in range(0,len(bos['OPEN_DT'])):
    if k not in rows:
        drop_rows.append(k)
feb_bos = bos.drop(drop_rows)
print feb_bos['OPEN_DT']
````

````python
# Filters the dataframe and keeps only the cases that were opened in February and are still open
rows = []
count = 0
for i in bos['OPEN_DT'].values[:]: 
    if not i == np.datetime64('NaT'):
        j = np.datetime_as_string(i, timezone='local')[:10]
        if j[5:7] == '02':
            if bos['CASE_STATUS'].values[count] == 'Open':
                rows.append(count)
    count += 1
drop_rows = []
for k in range(0,len(bos['OPEN_DT'])):
    if not (k in rows):
        drop_rows.append(k)
feb_bos = bos.drop(drop_rows)
print feb_bos['OPEN_DT']
````

## Step 3. Advanced filtering & joins

By now, you should have noticed that, for each day, there are multiple records in our dataframe. Use the `groupby` function to get, for each day, the counts for all the case categories. Plot the result for the top 5 categories as you see fit. **(20 pts)**

````python
# For each day, the counts for all the case categories. Plots the result for the top 5 categories
top_five = {}
counts = {}
days = bos.groupby([pd.Grouper(key = 'OPEN_DT', freq = '24h'), 'CASE_TITLE']) #group by both day and case_title
for key, value in days:
    if key[1] not in counts.keys():
        counts[key[1]] = value['CASE_TITLE'].count()
    else:
        counts[key[1]] += value['CASE_TITLE'].count()

case_count = counts.values()
titles = counts.keys()
numbers = np.array(case_count)        
indices = numbers.argsort()[::-1][:5]

for index in indices:
    top_five[titles[index]] = case_count[index]
    
    
five_cats=pd.DataFrame(list(top_five.itervalues()),columns=['Top Five Cases'],index=list(top_five.iterkeys()))
ax = five_cats.plot(kind='barh', title = 'Count of Top Five Case Categories', color = 'g')
ax.set_xlabel('Count')
ax.set_ylabel('Case Title')
````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
#Aggregates the measurements for the snow for every day in the dataset as the mean for each day
# in every station

snow_day = snow_data.groupby(pd.Grouper(key = 'DATE', freq = '24h'))

agg_snow = {}
count_per_day = {}
for key, value in snow_day:
    for i in value['SNOW'].values[:]:
        k = key.date()
        if k in agg_snow.keys():
            if i >= 0: #Ignore negative values as corrupted data
                agg_snow[k] += i
                count_per_day[k] += 1
        else:
            if i >= 0:
                agg_snow[k] = i
                count_per_day[k] = 1

agg_snow_days = []
for item in agg_snow.keys():
    agg_snow_days.append(item)
    agg_snow[item] = (agg_snow[item]/count_per_day[item])/10.0

agg_snow_days = sorted(agg_snow_days)
all_snow = []
for item in agg_snow_days:
    key = str(item)
    all_snow.append(agg_snow[item])    

colors = [0.9803921568627451, 0.5058823529411764, 0.4549019607843137]
all_the_snow =pd.DataFrame(all_snow,columns=['Amount of Snow'],index= agg_snow_days)
ax = all_the_snow.plot(kind='bar', title = 'Aggregate Snowfall per Day', color = colors)
ax.set_xlabel('Day')
ax.set_ylabel('Snowfall (cm)')
````

## Step 4. Matching trends

The last thing we will do is to join the snowfall dataframe with that of the top 5 categories of calls, per day. Don't forget to fill the NaN values with 0. You want to do an outer join (if there are calls for a day but no snowfall data, we want to create a 0 record). Save that to a dataframe called `calls_on_snow`. The first 5 columns should be the counts for the categories, and the 6th will be the snowfall data. **(10 pts)**

````python
#create a dataFrame called calls_on_snow where the rows are all the days and the columns are
# the top 5 categories. Do it for each day in the top 5 categories. 

call_days = []
col1 = {}
col2 = {}
col3 = {}
col4 = {}
col5 = {}
top_titles = top_five.keys()
for key, value in days:
    k = key[0].date()
    if key[1] in top_titles:
        if key[1] == top_titles[0]:
            if k in col1.keys():
                col1[k] += value['CASE_TITLE'].count()
            else:
                col1[k] = value['CASE_TITLE'].count()
        if key[1] == top_titles[1]:
            if k in col2.keys():
                col2[k] += value['CASE_TITLE'].count()
            else:
                col2[k] = value['CASE_TITLE'].count()
        if key[1] == top_titles[2]:
            if k in col3.keys():
                col3[k] += value['CASE_TITLE'].count()
            else:
                col3[k] = value['CASE_TITLE'].count()
        if key[1] == top_titles[3]:
            if k in col4.keys():
                col4[k] += value['CASE_TITLE'].count()
            else:
                col4[k] = value['CASE_TITLE'].count()
        if key[1] == top_titles[4]:
            if k in col5.keys():
                col5[k] += value['CASE_TITLE'].count()
            else:
                col5[k] = value['CASE_TITLE'].count()
    if k not in call_days:
        call_days.append(k)

call_days = sorted(call_days) #index by this for the new dataframe 

col1_sorted = []
col2_sorted = []
col3_sorted = []
col4_sorted = []
col5_sorted = []
col6 = []
for d in call_days:
    if d in agg_snow.keys():
        col6.append(agg_snow[d])
    else:
        col6.append(0)
    if d in col1.keys():
        col1_sorted.append(col1[d])
    else:
        col1_sorted.append(0)
    if d in col2.keys():
        col2_sorted.append(col2[d])
    else:
        col2_sorted.append(0)
    if d in col3.keys():
        col3_sorted.append(col3[d])
    else:
        col3_sorted.append(0)
    if d in col4.keys():
        col4_sorted.append(col4[d])
    else:
        col4_sorted.append(0)
    if d in col5.keys():
        col5_sorted.append(col5[d])
    else:
        col5_sorted.append(0)

calls_on_snow = pd.DataFrame(col1_sorted,columns=[top_titles[0]], index = call_days)
calls_on_snow[top_titles[1]] = col2_sorted
calls_on_snow[top_titles[2]] = col3_sorted
calls_on_snow[top_titles[3]] = col4_sorted
calls_on_snow[top_titles[4]] = col5_sorted
calls_on_snow['SNOW (cm)'] = col6
calls_on_snow
````

Now, you can show the connection of the two trends (the calls and the snow) in a plot. I provide some sample code, but you should also try your own versions.

````python
fig, axes = plt.subplots(nrows=2,figsize=(12,5), sharex=True)

calls_on_snow.ix[:,0:5].plot(ax=axes[0], color = ['b', 'g', 'r', 'm', 'y'])
calls_on_snow.ix[:,5].plot(ax=axes[1],label="snowfall (in cm)", color='c')

axes[0].set_title("The effect of the winter snowstorms on Boston's 311 Hotline for 2015")
axes[0].set_ylabel("# Calls")
axes[1].set_ylabel("Snowfall in cm")
axes[1].set_xlabel("Date")
axes[0].xaxis.grid(False)
axes[1].xaxis.grid(False)
````
