```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import sklearn.preprocessing as sk_preprocess
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import sklearn.metrics as metrics
```

```python
boston = pd.read_csv('../Homework-2/cases_2015.csv')
```

## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
```

```python
a = np.array(boston.LATITUDE)
b = np.array(boston.LONGITUDE)
mat = np.column_stack((a,b))
stnd = sk_preprocess.StandardScaler(with_mean = True)
m = stnd.fit_transform(mat)
#evaluate_clusters(m, 15) #Found the best number of clusters to be around 4 

kmeans = KMeans(init='k-means++', n_clusters=4, n_init=10, random_state = 1)
kmeans.fit_predict(m)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
c1 = []
c2 = []
c3 = []
c4 = []
#print len(kmeans.labels_)
for i in range(0, len(labels)):
    if(labels[i] == 0):
        c1.append([m[i][0], m[i][1]])
    if(labels[i] == 1):
        c2.append([m[i][0], m[i][1]])
    if(labels[i] ==2):
        c3.append([m[i][0], m[i][1]])
    if(labels[i]==3):
        c4.append([m[i][0], m[i][1]])

c1 = np.array(c1)
c2 = np.array(c2)
c3 = np.array(c3)
c4 = np.array(c4)
colors = ['b', 'g', 'r','m']
plt.scatter(centers[:,1], centers[:,0], color = colors,marker = '+', s = 30, alpha = 0.1)
plt.scatter(c1[:,1], c1[:,0], color = 'b',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c2[:,1], c2[:,0], color = 'g',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c3[:,1], c3[:,0], color = 'r',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c4[:,1], c4[:,0], color = 'm',s = 30,marker = '+', alpha = 0.1)
plt.scatter(centers[:,1], centers[:,0], marker = 'o', s=500, linewidths=2, c= 'none')
plt.scatter(centers[:,1], centers[:,0], marker = 'x', s=500, linewidths=2, c= 'k')
plt.show()
```

## Step 2. Using multiple features

The next step in our analysis is to use more features than just the geolocation of a record (whose clustering only gives us information about the neighborhoods, which should not be a huge revelation!) Filter the dataset to get more features, transform them as required and compute another clustering. You should argue about the quality of your clustering and use domain knowledge (i.e. the fact that you are locals) to explore the clusters you produced and present them to the reader. **(pts 30)**

```python
#cluster by geo-location and whether or not case is still open 
lats_open = []
longs_open = []
lats_closed = []
longs_closed = []
count = 0
for i in boston.CASE_STATUS:
    if i == 'Open':
        lats_open.append(boston.LATITUDE.values[count])
        longs_open.append(boston.LONGITUDE.values[count])
    
    count += 1

a_open = np.array(lats_open)
b_open = np.array(longs_open)
mat_open = np.column_stack((a_open,b_open))
stnd_open = sk_preprocess.StandardScaler(with_mean = True)
m_open = stnd.fit_transform(mat_open)

kmeans_open = KMeans(init='k-means++', n_clusters=4, n_init=10, random_state = 1)
kmeans_open.fit_predict(m_open)
centers_open = kmeans_open.cluster_centers_
labels_open = kmeans_open.labels_
c1_open = []
c2_open = []
c3_open = []
c4_open = []
for i in range(0, len(labels_open)):
    if(labels_open[i] == 0):
        c1_open.append([m_open[i][0], m_open[i][1]])
    if(labels_open[i] == 1):
        c2_open.append([m_open[i][0], m_open[i][1]])
    if(labels_open[i] ==2):
        c3_open.append([m_open[i][0], m_open[i][1]])
    if(labels_open[i]==3):
        c4_open.append([m_open[i][0], m_open[i][1]])

c1_open = np.array(c1_open)
c2_open = np.array(c2_open)
c3_open = np.array(c3_open)
c4_open = np.array(c4_open)
colors_open = ['b', 'g', 'r','m']
figure =plt.figure()
plt.scatter( centers_open[:,1], centers_open[:,0], color = colors_open,marker = '+', s = 30, alpha = 0.1)
plt.scatter(c1_open[:,1], c1_open[:,0], color = 'b',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c2_open[:,1], c2_open[:,0], color = 'g',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c3_open[:,1], c3_open[:,0], color = 'r',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c4_open[:,1], c4_open[:,0], color = 'm',s = 30,marker = '+', alpha = 0.1)
plt.scatter(centers_open[:,1], centers_open[:,0], marker = 'o', s=500, linewidths=2, c= 'none')
plt.scatter(centers_open[:,1], centers_open[:,0], marker = 'x', s=500, linewidths=2, c= 'k')
plt.title('OPEN CASES')
plt.show()

###############################################################################################
count = 0
for i in boston.CASE_STATUS:
    if i == 'Closed':
        lats_closed.append(boston.LATITUDE.values[count])
        longs_closed.append(boston.LONGITUDE.values[count])
    count += 1
a_closed = np.array(lats_closed)
b_closed = np.array(longs_closed)
mat_closed = np.column_stack((a_closed,b_closed))
stnd_closed = sk_preprocess.StandardScaler(with_mean = True)
m_closed = stnd.fit_transform(mat_closed)

kmeans_closed = KMeans(init='k-means++', n_clusters=4, n_init=10, random_state = 1)
kmeans_closed.fit_predict(m_closed)
centers_closed = kmeans_closed.cluster_centers_
labels_closed = kmeans_closed.labels_
c1_closed = []
c2_closed = []
c3_closed = []
c4_closed = []
for i in range(0, len(labels_closed)):
    if(labels_closed[i] == 0):
        c1_closed.append([m_closed[i][0], m_closed[i][1]])
    if(labels_closed[i] == 1):
        c2_closed.append([m_closed[i][0], m_closed[i][1]])
    if(labels_closed[i] ==2):
        c3_closed.append([m_closed[i][0], m_closed[i][1]])
    if(labels_closed[i]==3):
        c4_closed.append([m_closed[i][0], m_closed[i][1]])

c1_closed = np.array(c1_closed)
c2_closed = np.array(c2_closed)
c3_closed = np.array(c3_closed)
c4_closed = np.array(c4_closed)
colors_closed = ['g', 'r', 'b','m']
plt.scatter( centers_closed[:,1], centers_closed[:,0], color = colors_closed,marker = '+', s = 30, alpha = 0.1)
plt.scatter(c1_closed[:,1], c1_closed[:,0], color = 'g',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c2_closed[:,1], c2_closed[:,0], color = 'r',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c3_closed[:,1], c3_closed[:,0], color = 'b',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c4_closed[:,1], c4_closed[:,0], color = 'm',s = 30,marker = '+', alpha = 0.1)
plt.scatter(centers_closed[:,1], centers_closed[:,0], marker = 'o', s=500, linewidths=2, c= 'none')
plt.scatter(centers_closed[:,1], centers_closed[:,0], marker = 'x', s=500, linewidths=2, c= 'k')
plt.title('CLOSED CASES')
plt.show()

```
