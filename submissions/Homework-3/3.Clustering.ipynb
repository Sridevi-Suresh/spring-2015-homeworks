## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
#I use this function in order to figure out the optimum number of clusters. I use the kmeans
#intertia values that is returned by the scikit-learn kmeans function to plot my error vs the
#number of clusters. I pick the number of clusters to be the "knee" of the graph plotted. 

def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
```

```python
a = np.array(boston.LATITUDE)
b = np.array(boston.LONGITUDE)
mat = np.column_stack((a,b))
stnd = sk_preprocess.StandardScaler(with_mean = True)
m = stnd.fit_transform(mat)
#evaluate_clusters(m, 15) #Found the best number of clusters to be around 4 

kmeans = KMeans(init='k-means++', n_clusters=4, n_init=10, random_state = 1)
kmeans.fit_predict(m)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
c1 = []
c2 = []
c3 = []
c4 = []
#print len(kmeans.labels_)
for i in range(0, len(labels)):
    if(labels[i] == 0):
        c1.append([m[i][0], m[i][1]])
    if(labels[i] == 1):
        c2.append([m[i][0], m[i][1]])
    if(labels[i] ==2):
        c3.append([m[i][0], m[i][1]])
    if(labels[i]==3):
        c4.append([m[i][0], m[i][1]])

c1 = np.array(c1)
c2 = np.array(c2)
c3 = np.array(c3)
c4 = np.array(c4)
colors = ['b', 'g', 'r','m']
plt.scatter(centers[:,1], centers[:,0], color = colors,marker = '+', s = 30, alpha = 0.1)
plt.scatter(c1[:,1], c1[:,0], color = 'b',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c2[:,1], c2[:,0], color = 'g',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c3[:,1], c3[:,0], color = 'r',s = 30,marker = '+', alpha = 0.1)
plt.scatter(c4[:,1], c4[:,0], color = 'm',s = 30,marker = '+', alpha = 0.1)
plt.scatter(centers[:,1], centers[:,0], marker = 'o', s=500, linewidths=2, c= 'none')
plt.scatter(centers[:,1], centers[:,0], marker = 'x', s=500, linewidths=2, c= 'k')
plt.show()
```

```python
"""Explanation of my clustering above:

Many of the features of the data set are categorical so I found clustering with multiple features
a bit of a daunting task. I decided to consider simply the case status as this binarizes the
data. I grouped together the cases that are still open together and the cases that are closed
together. Then, I clustered (separately) the cases that are open by geolocation and the cases
that are closed by geolocation. Those are the two plots you see above.** The reason I decided to
cluster this way to see if there was any relation between the case status and geolocation. There
is not much to learn from this dataset but we can see that there are more closed cases than open
cases. As well, generally the north (color blue) area has had more cases over all. We can then 
see a ranking of the cases that are open and the cases that are closed be geolocation. If we use
some domain knowledge then we can see this clustering makes sense. This data set consists of the
calls made during the past two months which is when we experienced very heavy snow. The north 
area is a bit more suburb-like, hence it makes sense that they experienced more calls than other
areas over the past two months. 

**I chose two plots as this makes the result easier to visualize if I were to plot the clustering
on the same plot, then since there is overlap in the geolocations, it becomes very hard to 
discern the conclusion I made. 

"""
```

```python
#First give a score to each case's subject based off of my own importance related to the snow 
#storms Boston has experienced in the past couple months. Then, use the function evaluate clusters
#in order to figure out the optimum number of clusters to use. Run kmeans using that k value and
#then further cluster each clustering by geolocation and plot. 

#Scores: Snow - 5, Emergency - 6, Cleaning - 3, Plow - 3, Heat - 4, Traffic - 3, other - 1

score = [0]*len(boston.TYPE)
idx = 0
for t in boston.TYPE:
    other = False
    if t.find('Snow') >= 0:
        score[idx] += 5
        other = True
    if t.find('Emergency') >= 0:
        score[idx] += 6
        other = True 
    if t.find('Cleaning') >= 0:
        score[idx] += 3
        other = True
    if t.find('Plow') >= 0:
        score[idx] += 3
        other = True
    if t.find('Heat') >= 0:
        score[idx] += 4
        other = True
    if t.find('Traffic') >= 0:
        score[idx] += 3
        other = True
    if other == False:
        score[idx] += 1
    idx += 1

score = np.array(score)
data = np.column_stack([score])
#evaluate_clusters(data, 15) #good number of clusters is 3
k_means = KMeans(init='k-means++', n_clusters=3, n_init=10, random_state = 1)
k_means.fit_predict(data)
cents = k_means.cluster_centers_
labs = k_means.labels_
clust1 = []
clust2 = []
clust3 = []

a = np.array(boston.LATITUDE)
b = np.array(boston.LONGITUDE)
mat = np.column_stack((a,b))
stnd = sk_preprocess.StandardScaler(with_mean = True)
m = stnd.fit_transform(mat)

idx = 0
d1 = 100000
d2 = 100000
d3 = 100000
rep1 = []
rep2 = []
rep3 = []
for element in score:
    dist1 = abs(element - cents[0])
    dist2 = abs(element - cents[1])
    dist3 = abs(element - cents[2])
    if dist1 < dist2 and dist1 < dist3:
        if dist1 < d1:
            rep1 = [m[idx][0], m[idx][1]]
            d1 = dist1
    if dist2 < dist1 and dist2 < dist3:
        if dist2 < d2:
            rep2 = [m[idx][0], m[idx][1]]
            d2 = dist2
    if dist3 < dist1 and dist3 < dist2:
        if dist3 < d3:
            rep3 = [m[idx][0], m[idx][1]]
            d3 = dist3
    idx = idx + 1
    
for i in range(0, len(labs)):
    if(labs[i] == 0):
        clust1.append([m[i][0], m[i][1]])
    if(labs[i] == 1):
        clust2.append([m[i][0], m[i][1]])
    if(labs[i] ==2):
        clust3.append([m[i][0], m[i][1]])

cl1 = clust1
cl2 = clust2
cl3 = clust3
clust1 = np.array(clust1)
clust2 = np.array(clust2)
clust3 = np.array(clust3)

#colors g, k, m
l1 =c1.tolist()
l2 = c2.tolist()
l3 = c3.tolist()
l4 = c4.tolist()

c = ''
if rep1 in l1:
    c = 'b'
if rep1 in l2:
    c = 'g'
if rep1 in l3:
    c = 'r'
if rep1 in l4:
    c = 'm'
   
plt.scatter(rep1[1], rep1[0], color = c, marker = 'o', s = 50, alpha = 0.1)

coord1 = []
coord2 = []
coord3 = []
coord4 = []


for coord in cl1:
    if coord in l1:
        coord1.append(coord)
    if coord in l2:
        coord2.append(coord)
    if coord in l3:
        coord3.append(coord)
    if coord in l4:
        coord4.append(coord)

coord1 = np.array(coord1)
coord2 = np.array(coord2)
coord3 = np.array(coord3)
coord4 = np.array(coord4)
plt.scatter(coord1[:,1], coord1[:,0], color = 'b', marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord2[:,1], coord2[:,0], color = 'g', marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord3[:,1], coord3[:,0], color = 'm',marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord4[:,1], coord4[:,0], color = 'r', marker = '+', s = 50, alpha = 0.1)
plt.title('Low Score Cases')
plt.show()    


c = ''
if rep2 in l1:
    c = 'b'
if rep2 in l2:
    c = 'g'
if rep2 in l3:
    c = 'r'
if rep2 in l4:
    c = 'm'
   
plt.scatter(rep2[1], rep2[0], color = c, marker = 'o', s = 50, alpha = 0.1)

coord1 = []
coord2 = []
coord3 = []
coord4 = []

for coord in cl2:
    if coord in l1:
        coord1.append(coord)
    if coord in l2:
        coord2.append(coord)
    if coord in l3:
        coord3.append(coord)
    if coord in l4:
        coord4.append(coord)

coord1 = np.array(coord1)
coord2 = np.array(coord2)
coord3 = np.array(coord3)
coord4 = np.array(coord4)
plt.scatter(coord1[:,1], coord1[:,0], color = 'b', marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord2[:,1], coord2[:,0], color = 'g', marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord3[:,1], coord3[:,0], color = 'm',marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord4[:,1], coord4[:,0], color = 'r', marker = '+', s = 50, alpha = 0.1)
plt.title('Medium Score Cases')
plt.show()  

c = ''
if rep3 in l1:
    c = 'b'
if rep3 in l2:
    c = 'g'
if rep3 in l3:
    c = 'r'
if rep3 in l4:
    c = 'm'
   
plt.scatter(rep3[1], rep3[0], color = c, marker = 'o', s = 50, alpha = 0.1)

coord1 = []
coord2 = []
coord3 = []
coord4 = []

for coord in cl3:
    if coord in l1:
        coord1.append(coord)
    if coord in l2:
        coord2.append(coord)
    if coord in l3:
        coord3.append(coord)
    if coord in l4:
        coord4.append(coord)

coord1 = np.array(coord1)
coord2 = np.array(coord2)
coord3 = np.array(coord3)
coord4 = np.array(coord4)
plt.scatter(coord1[:,1], coord1[:,0], color = 'b', marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord2[:,1], coord2[:,0], color = 'g', marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord3[:,1], coord3[:,0], color = 'm',marker = '+', s = 50, alpha = 0.1)
plt.scatter(coord4[:,1], coord4[:,0], color = 'r', marker = '+', s = 50, alpha = 0.1)
plt.title('High Score Cases')
plt.show() 
```
