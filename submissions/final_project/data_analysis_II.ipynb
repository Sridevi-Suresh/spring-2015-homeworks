```python
#Alec Larsen - University of the Witwatersrand, South Africa, 2012 import shlex, subprocess

#This is used to call and retrieve the negative sentiment score from the sentistrength program.

#Input: a string to pass into the sentistrength program.
#Output: a score from -4 - 0 depending on how negative the string is. 
import subprocess
import shlex
def RateSentimentNeg(sentiString):
    #open a subprocess using shlex to get the command line string into the correct args list format
    p = subprocess.Popen(shlex.split("java -jar SentiStrength.jar stdin sentidata C:/Users/Sridevi/Documents/Computer_Science/cs591DS/project/SentStrength/"),stdin=subprocess.PIPE,stdout=subprocess.PIPE,stderr=subprocess.PIPE)
    #communicate via stdin the string to be rated. Note that all spaces are replaced with +
    stdout_text, stderr_text = p.communicate(sentiString.replace(" ","+"))
    #remove the tab spacing between the positive and negative ratings. e.g. 1    -5 -> 1-5
    stdout_text = stdout_text.rstrip().replace("\t",",").split(",")
    return int(stdout_text[1]) + 1
```

```python
#Alec Larsen - University of the Witwatersrand, South Africa, 2012 import shlex, subprocess

#This is used to call and retrieve the positive sentiment score from the sentistrength program.

#Input: a string to pass into the sentistrength program.
#Output: a score from 0 - 4 depending on how positive the string is. 

import subprocess
import shlex
def RateSentimentPos(sentiString):
    #open a subprocess using shlex to get the command line string into the correct args list format
    p = subprocess.Popen(shlex.split("java -jar SentiStrength.jar stdin sentidata C:/Users/Sridevi/Documents/Computer_Science/cs591DS/project/SentStrength/"),stdin=subprocess.PIPE,stdout=subprocess.PIPE,stderr=subprocess.PIPE)
    #communicate via stdin the string to be rated. Note that all spaces are replaced with +
    stdout_text, stderr_text = p.communicate(sentiString.replace(" ","+"))
    #remove the tab spacing between the positive and negative ratings. e.g. 1    -5 -> 1-5
    stdout_text = stdout_text.rstrip().replace("\t",",").split(",")
    return int(stdout_text[0]) - 1
```

```python
#Read in specifically the captions, comments, and number of comments from each park.

import csv 

dl_captions = disneyland.caption
dw_captions = disneyworld.caption
dl_score = []
dw_score = []

with open('disneyland_comments.csv', 'rb') as f:
    dl_comments = list(csv.reader(f))
f.close()

with open('disneyworld_comments.csv', 'rb') as f:
    dw_comments = list(csv.reader(f))
f.close()

dl_numcomments = disneyland.num_comments
dw_numcomments = disneyworld.num_comments
```

```python
#Retrieve the negative sentiment scores by calling the RateSentimentNeg function. Then, use pickle to save the scores from each park. Not 
#only add the score of the caption but also the score of each of the comments to the caption in order to get an overall score for a post. 

idx = 0
for item in dl_captions:
    item = item.replace("#", " ")
    item = item.replace("@", " ")
    score = RateSentimentNeg(item)
    for i in range(0,dl_numcomments[idx]):
        com = dl_comments[idx+1][i]
        com = com.replace("#", " ")
        com = com.replace("@", " ")
        if com:
            score += RateSentimentNeg(com)
    dl_score.append(score)
    idx +=1
    
idx = 0
for item in dw_captions:
    item = item.replace("#", " ")
    item = item.replace("@", " ")
    score = RateSentimentNeg(item)
    for i in range(0,dw_numcomments[idx]):
        com = dw_comments[idx+1][i]
        com = com.replace("#", " ")
        com = com.replace("@", " ")
        if com:
            score += RateSentimentNeg(com)
    dw_score.append(score)
    idx +=1

pickle.dump(dl_score, open("dl_negscore.p", "wb"))
pickle.dump(dw_score,open("dw_negscore.p","wb"))
```

```python
#Read in all of the scores from each of the individual parks and plot a graph to look at the frequency of negative scores for each park. 

dl_negscores = pickle.load(open("dl_negscore.p", "rb"))
dw_negscores = pickle.load(open("dw_negscore.p", "rb"))

dl_notfun = {}
dw_notfun= {}

for item in dl_negscores:
    if item not in dl_notfun.keys():
        dl_notfun[item] = 1
    else:
        dl_notfun[item] += 1
        
for item in dw_negscores:
    if item not in dw_notfun.keys():
        dw_notfun[item] = 1
    else:
        dw_notfun[item] += 1
        
for k in dl_notfun.keys():
    if k not in dw_notfun.keys():
        dw_notfun[k] = 0

for k in dw_notfun.keys():
    if k not in dl_notfun.keys():
        dl_notfun[k] = 0
    
mini = min(min(dl_notfun), min(dw_notfun))
nums = range(mini,0)
for n in nums:
    if n not in dl_notfun.keys():
        dl_notfun[n] = 0
    if n not in dw_notfun.keys():
        dw_notfun[n] = 0

dl_nf = sorted(dl_notfun)
dw_nf = sorted(dw_notfun)

dl_vals = []
dw_vals = []

for item in dl_nf:
    dl_vals.append(dl_notfun[item])

for item in dw_nf:
    dw_vals.append(dw_notfun[item])

notfun = pd.DataFrame(dl_vals, columns=["Disneyland"], index = dl_nf)
notfun["Disneyworld"] = dw_vals

ax = notfun.plot(kind='bar', title = 'Negative Sentiment Scores\n for Disneyland and DisneyWorld',figsize=(10, 6)
                 ,color = ['#00FF00', '#8B4513'], rot=45)
ax.set_xlabel("Negative Score",fontsize=15)
ax.set_ylabel("Count of Score",fontsize=15)
ax.grid(b=False)
pylab.savefig('neg_sentiment.png',bbox_inches='tight')
```

```python
#Create a dictionary for each park where the key is a subpark and then go through the captions and add it to a list of values per subpark

dl_parks = {'adventureland':[], 'tomorrowland':[], 'crittercountry':[], 'neworleanssquare':[], 'toontown': [], 'mainstreet': [], 'frontierland': []}
dw_parks = {'downtowndisney':[], 'hollywoodstudios':[], 'magickingdom':[], 'epcot':[], 'animalkingdom':[]}


for c in dl_captions:
    for key in dl_parks.keys():
        if key in c:
            dl_parks[key].append(c)
    

for c in dw_captions:
    for key in dw_parks.keys():
        if key in c:
            dw_parks[key].append(c)
```

```python
#Use the two functions RateSentimentPos and RateSentimentNeg in order to keep track of the positive scores for captions and comments in 
#each of the subparks of a park. Then, save all the scores using pickle. 

dl_park_posscore = {'adventureland':[], 'tomorrowland':[], 'crittercountry':[], 'neworleanssquare':[], 'toontown': [], 'mainstreet': [], 'frontierland': []}
dw_park_posscore = {'downtowndisney':[], 'hollywoodstudios':[], 'magickingdom':[], 'epcot':[], 'animalkingdom':[]}

dl_park_negscore = {'adventureland':[], 'tomorrowland':[], 'crittercountry':[], 'neworleanssquare':[], 'toontown': [], 'mainstreet': [], 'frontierland': []}
dw_park_negscore = {'downtowndisney':[], 'hollywoodstudios':[], 'magickingdom':[], 'epcot':[], 'animalkingdom':[]}

idx = 0
for k in dl_parks:
    for val in dl_parks[k]:
        val = val.replace("#", " ")
        val = val.replace("@", " ")
        posscore = RateSentimentPos(val)
        negscore = RateSentimentNeg(val)
        for i in range(0,dl_numcomments[idx]):
            com = dl_comments[idx+1][i]
            com = com.replace("#", " ")
            com = com.replace("@", " ")
            if com:
                posscore += RateSentimentPos(com)
                negscore += RateSentimentNeg(com)
        dl_park_posscore[k].append(posscore)
        dl_park_negscore[k].append(negscore)
    idx += 1
        
idx = 0
for k in dw_parks:
    for val in dw_parks[k]:
        val = val.replace("#", " ")
        val = val.replace("@", " ")
        posscore = RateSentimentPos(val)
        negscore = RateSentimentNeg(val)
        for i in range(0,dw_numcomments[idx]):
            com = dw_comments[idx+1][i]
            com = com.replace("#", " ")
            com = com.replace("@", " ")
            if com:
                posscore += RateSentimentPos(com)
                negscore += RateSentimentNeg(com)
        dw_park_posscore[k].append(posscore)
        dw_park_negscore[k].append(negscore)
    idx += 1
        
pickle.dump(dl_park_posscore, open("dl_park_posscore.p", "wb"))
pickle.dump(dw_park_posscore,open("dw_park_posscore.p","wb"))
pickle.dump(dl_park_negscore, open("dl_park_negscore.p", "wb"))
pickle.dump(dw_park_negscore,open("dw_park_negscore.p","wb"))
```
