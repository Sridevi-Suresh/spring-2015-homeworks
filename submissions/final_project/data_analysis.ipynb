```python
#Alec Larsen - University of the Witwatersrand, South Africa, 2012 import shlex, subprocess
#This is used to call and retrieve the positive sentiment score from the sentistrength program.

#Input: a string to pass into the sentistrength program.
#Output: a score from 0 - 4 depending on how positive the string is. 
import subprocess
import shlex
def RateSentiment(sentiString):
    #open a subprocess using shlex to get the command line string into the correct args list format
    p = subprocess.Popen(shlex.split("java -jar SentiStrength.jar stdin sentidata C:/Users/Sridevi/Documents/Computer_Science/cs591DS/project/SentStrength/"),stdin=subprocess.PIPE,stdout=subprocess.PIPE,stderr=subprocess.PIPE)
    #communicate via stdin the string to be rated. Note that all spaces are replaced with +
    stdout_text, stderr_text = p.communicate(sentiString.replace(" ","+"))
    #remove the tab spacing between the positive and negative ratings. e.g. 1    -5 -> 1-5
    stdout_text = stdout_text.rstrip().replace("\t","")
    return int(stdout_text[0]) - 1
```

```python
#Read in specifically the captions, comments, and number of comments from each park.

import csv 

dl_captions = disneyland.caption
dw_captions = disneyworld.caption
dl_score = []
dw_score = []

with open('disneyland_comments.csv', 'rb') as f:
    dl_comments = list(csv.reader(f))
f.close()

with open('disneyworld_comments.csv', 'rb') as f:
    dw_comments = list(csv.reader(f))
f.close()

dl_numcomments = disneyland.num_comments
dw_numcomments = disneyworld.num_comments
```

```python
#Retrieve the positive sentiment scores by calling the RateSentiment function. Then, use pickle to save the scores from each park. Not only
#add the score of the caption but also the score of each of the comments to the caption in order to get an overall score for a post. 

idx = 0
for item in dl_captions:
    item = item.replace("#", " ")
    item = item.replace("@", " ")
    score = RateSentiment(item)
    for i in range(0,dl_numcomments[idx]):
        com = dl_comments[idx+1][i]
        com = com.replace("#", " ")
        com = com.replace("@", " ")
        if com:
            score += RateSentiment(com)
    dl_score.append(score)
    idx +=1
    
idx = 0
for item in dw_captions:
    item = item.replace("#", " ")
    item = item.replace("@", " ")
    score = RateSentiment(item)
    for i in range(0,dw_numcomments[idx]):
        com = dw_comments[idx+1][i]
        com = com.replace("#", " ")
        com = com.replace("@", " ")
        if com:
            score += RateSentiment(com)
    dw_score.append(score)
    idx +=1

pickle.dump(dl_score, open("dl_score.p", "wb"))
pickle.dump(dw_score,open("dw_score.p","wb"))
```

```python
#Read in all of the scores from each of the individual parks and plot a graph to look at the frequency of positive scores for each park. 

dl_scores = pickle.load(open("dl_score.p", "rb"))
dw_scores = pickle.load(open("dw_score.p", "rb"))

dl_fun = {}
dw_fun= {}

for item in dl_scores:
    if item not in dl_fun.keys():
        dl_fun[item] = 1
    else:
        dl_fun[item] += 1
        
for item in dw_scores:
    if item not in dw_fun.keys():
        dw_fun[item] = 1
    else:
        dw_fun[item] += 1

for k in dl_fun.keys():
    if k not in dw_fun.keys():
        dw_fun[k] = 0

for k in dw_fun.keys():
    if k not in dl_fun.keys():
        dl_fun[k] = 0
        
nums = range(0,22)
for n in nums:
    if n not in dl_fun.keys():
        dl_fun[n] = 0
    if n not in dw_fun.keys():
        dw_fun[n] = 0

fun = pd.DataFrame(dl_fun.values(), columns=["Disneyland"], index = dl_fun.keys())
fun["Disneyworld"] = dw_fun.values()

ax = fun.plot(kind='bar', title = 'Positive Sentiment Scores\n for Disneyland and DisneyWorld',figsize=(10, 6),color = ['#A52A2A', '#CD853F'], rot=45)
ax.set_xlabel("Positive Score",fontsize=15)
ax.set_ylabel("Count of Score",fontsize=15)
ax.grid(b=False)
pylab.savefig('pos_sentiment.png',bbox_inches='tight')
```
